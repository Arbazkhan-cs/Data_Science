<h1 style="color: red">Linear Regration with one variable</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# size = np.array([100, 200, 250, 300, 325, 500, 525, 600, 700, 850, 980]) # Input Variable: Feature
# prices = np.array([2000, 3000, 2500, 4000, 4700, 4500, 5500, 6000, 6200, 7000, 8000]) # Output Variable: Target

# Now getting the data from the csv file
df = pd.read_csv("house_prices.csv")
df.head()
df.drop_duplicates(inplace=True)
area = df.Area
prices = df.Price
```

```python
plt.scatter(area, prices, color="green")
plt.show()
```

```python
# T0 create a linear regrestion model for this than first we have to find our m and b variables
area_mean = np.mean(area)
prices_mean = np.mean(prices)
```

```python
# To find numerator we have to find the sum of E(x-x_mean)*(y-y_mean)
# To find denominator we have to find the sum of E((x-x_mean)**2) 
numerator = np.sum((area-area_mean)*(prices-prices_mean))
denominator = np.sum((area-area_mean)**2)
```

```python
# slop = numerator/denominator and intercept = mean_y-x_mean*slop
m = numerator/denominator
b = prices_mean - m*area_mean
m, b
```

```python
# Now we can get the all predicted value by suing the formula: y = mx+b
regressionLine = m*area+b
regressionLine.head()
```

```python
# Now ploting our pridicted linear regration line to our dataset
plt.scatter(area, prices, color="green")
plt.plot(area, regressionLine, color="blue")
```

```python
# Now making our graph more enhanced by using labels
plt.xlabel("area (sqr feet)")
plt.ylabel("prices (US dollar)")
plt.title("House Price Prediction")
plt.scatter(area, prices, color="green", label="Prices")
plt.plot(area, regressionLine, color="blue", label="regretion line")
plt.legend()
plt.grid()
```

```python
# Now creating a function to predict the prices according to the size of the area
def predict(size):
    return m * size + b

predict(1500) # the price of 1500sq feet is around = 2,46,950
predict(2000) # 3,61,260
predict(2500) # 4,75,563
predict(1600)
```

```python
# Now the mean squared error in our regrestion should be
length = len(prices)
MSE = np.mean((prices - regressionLine) ** 2)
MSE
```

```python
# Now finding our R Squared to see the goodness of our eqn
SSR = np.sum((regressionLine - prices)**2)
SST = np.sum((prices - prices_mean)**2)
Rsqr = 1 - (SSR / SST)
Rsqr # Range: [1-0]
```
