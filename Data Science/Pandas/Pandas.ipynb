<h3 style="color: red">Working With Excel File</h3>

```python
df = pd.read_excel('stock.xlsx', sheet_name="Sheet1")
df
```

```python
# converters
def convert_people_cell(cell):
    if cell == 'n.a.':
        return "Unknown"
    return cell

def convert_revenue_cell(cell):
    if cell == -1:
        return 0
    return cell

# converters are used to use some functions while fetching the data
df = pd.read_excel('stock.xlsx', sheet_name="Sheet1", converters={
    'people': convert_people_cell,
    'revenue': convert_revenue_cell
}) # na_value=['n.a.', 'not availabe'] this cannot be work in excel files you can create a converter for it
df
```

```python
df.to_excel("newexcel.xlsx", sheet_name="Stock") # this is how we can create another excel file
df.to_excel('newexcel.xlsx', sheet_name="Stock", startrow=1, startcol=1) # this will add the content after 1st row and col
```

<h2 style="color:blue">Chapter 5: Handling Missing Values By replace() function</h2>

```python
%%HTML
<iframe width="560" height="315" src="https://www.youtube.com/embed/XOxABiMhG2U" title="YouTube video player" frameborder="0" allow="accelerometer; autoplay; clipboard-write; encrypted-media; gyroscope; picture-in-picture; web-share" allowfullscreen></iframe>
```

```python
import pandas as pd
```

```python
data = {
    'day': ['1/1/2017', '1/2/2017', '1/3/2017', '1/4/2017', '1/5/2017', '1/6/2017', '1/7/2017'],
    'temperature': [32, -999999, 28, -999999, 32, 31, 34],
    'windspeed': [6, 7, -999999, 7, -999999, 2, 5],
    'event': ['Rain', 'Sunny', 'Snow', 'No Event', 'Rain', 'Sunny', 'No Event']
}

df = pd.DataFrame(data)
df
```

```python
# Here We have some special data(-999999) which is like an null value
# so we can use our replace function to replace this value by another value
new_df = df.replace(-999999, 'NaN') # np.NaN for numpy
new_df
```

```python
df['windspeed'][4] = -888888 # changing the value
df
```

```python
# Now we have 2 irrivelent values so instead of passing one value we can pass a list
new_df = df.replace([-999999, -888888], 'NaN')
new_df
```

```python
# we can also do for specific columns by dictionary
new_df = df.replace({
    'temperature' : -999999,
    'windspeed': [-999999, -888888],
    'event': 'No Event'
}, 'NaN')
new_df
```

```python
# we can also do by assigning irrevelent values to revelent values in dictionary
new_df = df.replace({
    -999999: 'NaN',
    -888888: 0,
    'No Event': 'Sunny'
})
new_df
```

```python
df['windspeed'][0] = '6 mps'
df['temperature'][4] = '32 F'
df['temperature'][1] = '32 C'
df
```

```python
# As now we have the some data with there units, so how we will remove these units?
# we can do this by using regex(regular Expression)
new_df = df.replace('[a-zA-Z]', '', regex=True) 
new_df
```

```python
# As we can see that this method is also deleted our alpha data
# so we can do this on some specific column
new_df = df.replace({
    'temperature': '[A-Za-z]',
    'windspeed': '[A-Za-z]'
}, '', regex=True)
new_df
```

```python
# Important is that we can also change the data form one list to another list
# so first let create one more data set
data = {
    'score': ['poor', 'average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'saif', 'imran', 'arbaz', 'ravi', 'romi']
}
df = pd.DataFrame(data)
df
```

```python
# if we want to change the list of this score to numeric, we can do this
new_df = df.replace(['poor', 'average', 'good', 'exceptional'], [1, 2, 3, 4])
new_df
```

<h2 style="color: blue">Chapter 6: Using Group By In Pandas</h2>

```python
%%HTML
<iframe width="560" height="315" src="https://www.youtube.com/embed/Wb2Tp35dZ-I" title="YouTube video player" frameborder="0" allow="accelerometer; autoplay; clipboard-write; encrypted-media; gyroscope; picture-in-picture; web-share" allowfullscreen></iframe>
```

```python
import pandas as pd
```

```python
data = {
    'day': ['1/1/2017', '1/2/2017', '1/3/2017', '1/4/2017', '1/5/2017', '1/6/2017', '1/7/2017','1/8/2017', '1/9/2017', '1/10/2017' ],
    'city': ['new york', 'new york', 'new york','new york', 'mumbai', 'mumbai', 'mumbai', 'paris', 'paris', 'paris'],
    'temperature': [32, 33, 28, 30, 32, 31, 34, 32, 28, 27],
    'windspeed': [6, 7, 5, 7, 3, 2, 5, 7, 6, 6],
    'event': ['Rain', 'Sunny', 'Snow', 'Snow', 'Rain', 'Sunny', 'Fog', 'Fog', 'Snow', 'Rain']
}
df = pd.DataFrame(data)
df
```

```python
# If we want to find maximum temperature in each of the cities, how we will do that?
# of if we want to find average wind speed per city. then how we will do this?

# So In this case we have a groupby function which makes it easy
g = df.groupby('city')
g # here we can see that it is a DataFrameGroupBy Object
```

```python
# so to work with this we can use a loop like this:
for city, city_df in g:
    print(city)
    print(city_df)  
    print()
    
# Here we can clearly see that the group by function created three differnet DataFrame for differnt city's
```

```python
# Now we can easily perform the operatiomns on it to solve our given querys
#Q1: Find maximum temperature in each of the cities
g.max()['temperature']
```

```python
#Q2: Find average wind speed per city
g.mean()['windspeed']
```

```python
# We can also do this
g.describe()
```

```python
# if we want to fetch a specific group we can use below method:
g.get_group('mumbai')
```

```python
# we can also plot a graph using metplotlib
%matplotlib inline
g.plot()
```

<h2 style="color: blue">Chapter 7: DataFrames Concat in Pandas</h2>

```python
%%HTML
<iframe width="560" height="315" src="https://www.youtube.com/embed/WGOEFok1szA" title="YouTube video player" frameborder="0" allow="accelerometer; autoplay; clipboard-write; encrypted-media; gyroscope; picture-in-picture; web-share" allowfullscreen></iframe>
```

```python
import pandas as pd
```

```python
# First DataFrame
india_weather = pd.DataFrame({
    'city': ['mumbai', 'delhi', 'banglore'],
    'humidity': [80, 60, 78],
    'temperature': [32, 45, 30]
})
india_weather
```

```python
# Second DataFrame
us_weather = pd.DataFrame({
    'city': ['new york', 'chicago', 'oriando'],
    'humidity': [68, 65, 75],
    'temperature': [21, 14, 35]
})
us_weather
```

```python
# Now if we want to combine these 2 data frames we con do it by using concat fucntion
df = pd.concat([us_weather, india_weather])
df
```

```python
# But in above we can see that the index are repeating based on there own index
# to solve this problem we can use a attribute ignore_index
df = pd.concat([us_weather, india_weather], ignore_index=True)
df
```

```python
# now here how we will discreminant the india city with us for tha n we can assign specifik keys to them
df = pd.concat([us_weather, india_weather], ignore_index=True, keys=["india", "us"])
df
```

```python
# we can see it is not working because as ignore_index=True which interupt this method so:
df = pd.concat([us_weather, india_weather], keys=['us', 'india'])
df
```

```python
df.loc["india"] # now we can get only the india data
```

```python
df = pd.concat([us_weather, india_weather]) # form this the rows are came after one and other
df
```

```python
# but what if we want to concat 2 string column by column, then we can use axis attribute
df = pd.concat([us_weather, india_weather], axis=1)
df
```

```python
# Last but not the least we have series also
# we can create a series uding Series method
s = pd.Series(['cloud', 'sunny', 'rain'], name="event")
s
```

```python
# now we can add this to our Data Frame
df = pd.concat([india_weather, s], axis=1)
df
```
