```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Sigmoid Function
def sigmoid(z):
    gz = 1 / (1 + np.exp(-z))
    return gz

# Gradient Descent For Logistic Regression
def gradient_descent(X, Y, W, b, alpha, iterations):
    m = len(X)
    
    i = 0
    totalCost = []
    while i < iterations:
        z = np.dot(X, W) + b
        f_wb = sigmoid(z)
        
        cost = (-1/m) * np.sum(Y*np.log(f_wb) + (1-Y)*np.log(1-f_wb))
        print(f"cost = {cost}, i = {i}")
        
        dw = (1/m) * np.dot(X.T, (f_wb - Y))
        db = (1/m) * np.sum(f_wb - Y)
        
        W = W - alpha * dw
        b = b - alpha * db
        
        totalCost.append(cost)
        i+=1
    
    return W, b, i, totalCost
```

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])  #(m,n)
Y = np.array([0, 0, 0, 1, 1, 1])  # n

print("X data:\n", X)
print("Y data:\n", Y)
```

```python
W = np.zeros(X.shape[1])
b = 0
alpha = 0.1
iterations = 1000

# Call gradient descent function
W_updated, b_updated, iteration, cost = gradient_descent(X, Y, W, b, alpha, iterations)

# Print updated weights and bias
print("Updated weights:", W_updated)
print("Updated bias:", b_updated)
```

```python
def predict(X, W, b):
    z = np.dot(X, W)+b
    prob = sigmoid(z)
    return np.round(prob)

Y_pred = predict(X, W_updated, b_updated)
Y_pred, Y
```
