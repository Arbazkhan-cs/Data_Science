```python
%%HTML
<iframe width="560" height="315" src="https://www.youtube.com/embed/oETDriX9n1w" title="YouTube video player" frameborder="0" allow="accelerometer; autoplay; clipboard-write; encrypted-media; gyroscope; picture-in-picture; web-share" allowfullscreen></iframe>
```

```python
import matplotlib.pyplot as plt
```

```python
days = [1, 2, 3, 4, 5, 6, 7]
max_temp = [50, 51, 52, 48, 47, 49, 46]
min_temp = [43, 42, 40, 44, 33, 35, 37]
avg_temp = [45, 48, 48, 46, 40, 42, 41]
```

```python
# we can plot all the temperature graph in one graph like this
plt.plot(days, max_temp)
plt.plot(days, min_temp)
plt.plot(days, avg_temp)
```

```python
# But in above graph we cannot see that which axis is describing which type of data
# So to label we have the labels method which are discussed in previous chapters
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("weather")

plt.plot(days, max_temp)
plt.plot(days, min_temp)
plt.plot(days, avg_temp)
```

```python
# Now are graphs is better
# but still now there is something issue like we don't now what these lines are representing or doing
# So to do that we have a legend property
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("weather")

plt.plot(days, max_temp, label="Max") # before legend property we have to define the labels on it
plt.plot(days, min_temp, label="Min")
plt.plot(days, avg_temp, label="Avg")

plt.legend()
# Now we can see on right corner of the graph that which line is representing which type of temperature
```

```python
# if we want legend on oth location than we can also do that by loc prperty
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("weather")

plt.plot(days, max_temp, label="Max") 
plt.plot(days, min_temp, label="Min")
plt.plot(days, avg_temp, label="Avg")

plt.legend(loc="lower left", shadow=True) # this will create a nice shadow on the legend
# We can use loc like:
# 'best': which is more suitable for the graph
# 'upper right'
# 'upper left'
# 'lower right'
# 'lower left'
# 'right'
# 'center left'
# 'center right'
# 'lower center'
# 'upper center'
# 'center'
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("weather")

plt.plot(days, max_temp, label="Max") 
plt.plot(days, min_temp, label="Min")
plt.plot(days, avg_temp, label="Avg")

plt.legend(loc="best", shadow=True, fontsize='large') # We can also change the font size
plt.grid() # this will create grid lines which lookes more cool
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
company = ['Google', 'Amazon', 'MSFT', 'Facebook']
revenue = [90, 136, 89, 27]
```

```python
# This is how we can create bar chart
plt.bar(company, revenue)
```

```python
revenue = [90, 136, 89, 27]
idx = np.arange(1, 5)

# here we have the x-axis in numerics
plt.bar(idx, revenue)
```

```python
# but what is we want to change this by a list of names
# so we can do this by using xticks function
plt.xticks(idx, company)
plt.bar(idx, revenue)
```

```python
# now let fromat our bar chart
plt.xlabel("Company's")
plt.ylabel("revenue")
plt.title("Comparision")

plt.bar(company, revenue, label="Revenue", width=0.4)
plt.legend()
```

```python
# Now we also have a list of profit in each company:
profit = [40, 10, 33, 20]
```

```python
# Let's add profit also
plt.xlabel("Company's")
plt.ylabel("revenue")
plt.title("Comparision")

plt.bar(company, revenue, label="Revenue", width=0.4)
plt.bar(company, profit, label="Profit", width=0.4)
plt.legend()
```

```python
# It's look great but we don't expect the graph like this
# we think a graph like side by side bar
# so to do that we have to change the company aria for both of this bars
# but company is a string type so that is why we create the idx for this
plt.xticks(idx, company)
plt.xlabel("Company's")
plt.ylabel("revenue")
plt.title("Comparision")

plt.bar(idx-0.2, revenue, label="Revenue", width=0.4)
plt.bar(idx+0.2, profit, label="Profit", width=0.4)
plt.legend()
# So now we have are fullfledge bar chart for the comparision
```

```python
# ***Note: Always use numeric data for x-axis and then you can change it by using xticks method.
```

```python
# We can also change the direction of the bar to horizontal bars: use plt.barh()
plt.yticks(idx, company)
plt.ylabel("Company's")
plt.xlabel("revenue")
plt.title("Comparision")

plt.barh(idx-0.2, revenue, label="Revenue")
plt.barh(idx+0.2, profit, label="Profit")
plt.legend()

# in this the width attribute doesn't exist
```

```python
import matplotlib.pyplot as plt
```

```python
# To create a histograph we can use hist function
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar)
```

```python
# We can also specifies how many pillers or bins we want 
plt.hist(blood_sugar, bins=3)
```

```python
# here all pillers are attech to each other so we can give some type of speces also
plt.hist(blood_sugar, bins=3, rwidth=0.95) # it's range lies btw [0, 1]
```

```python
# in above graph the x-axis are defined auto maticaly but we can define them according to us by changing the bins
plt.hist(blood_sugar, bins=[80, 100, 125, 150], rwidth=0.95)
```

```python
# we can also change the color
plt.hist(blood_sugar, bins=[80, 100, 125, 150], rwidth=0.95, color="g")
```

```python
# we can also create a step type graph like this
plt.hist(blood_sugar, bins=[80, 100, 125, 150], rwidth=0.95, color="g", histtype="step")
```

```python
# we can also define two graphs
blood_sugar_man = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_woman = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.hist([blood_sugar_man, blood_sugar_woman], bins=[80, 100, 125, 150], rwidth=0.95, color=["green", "orange"])
```

```python
# Now let's make it more effecient by including labels
plt.xlabel("Sugar Range")
plt.ylabel("Total Number Of Patients")
plt.title("Blood Sugar Anaylasis")

plt.hist([blood_sugar_man, blood_sugar_woman], bins=[80, 100, 125, 150], rwidth=0.95, color=["green", "orange"], 
         label=["Man", "Woman"])
plt.legend()
```

```python
# we can also change it's orientation to horizontal
plt.xlabel("Sugar Range")
plt.ylabel("Total Number Of Patients")
plt.title("Blood Sugar Anaylasis")

plt.hist([blood_sugar_man, blood_sugar_woman], bins=[80, 100, 125, 150], rwidth=0.95, 
         color=["green", "orange"], label=["Man", "Woman"], orientation="horizontal")
plt.legend()
```

```python
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone/internet Bill", "Car Rent", "Other Utllities"]
```

```python
# suppose we have data like above so, for these data the most suitable abd understandable chart is pie chart
plt.pie(exp_vals, labels=exp_labels)
```

```python
# but in above chart there are so many information which is not required for user so we can remove it by using show()

plt.pie(exp_vals, labels=exp_labels)
plt.show()
```

```python
# now it looks cool
# we can also adjust it's size by radius parameter
plt.pie(exp_vals, labels=exp_labels, radius=0.8)
plt.show()
```

```python
# but in above chart we can't see our percentages of each labels. So to do that we can use autopct parameter
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.2f%%") # %0.2f%%: it refers 2 points of decimal with percentage
plt.show()
```

```python
# we also have shadow property
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.2f%%", shadow="True")
plt.show 
```

```python
# we can also explode the pie's of the chart
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.2f%%", explode=[0, 0.1, 0, 0, 0], shadow=True)
plt.show()
```

```python
# we can also chage the direaction of the chart
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.2f%%", explode=[0, 0.1, 0, 0, 0], startangle=90)
plt.show()
```
