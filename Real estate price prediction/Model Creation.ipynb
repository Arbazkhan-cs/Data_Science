```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("final Data Set.csv")
```

```python
# Now let first convert our location to numeric values by one hot encoding method
dummies = pd.get_dummies(df1["location"])
dummies
```

```python
# Now concat the dummies with our df
# And to avoid dummies problem we will drop our other categry and represent other categry by all 0's
df2 = pd.concat([df1, dummies.drop("other", axis="columns")], axis="columns")
```

```python
df2.head()
```

```python
# Now we ca drop our location column
df3 = df2.drop("location", axis="columns")
df3.head()
```

<h1 style="color: blue">Model Building</h1>

```python
df3.shape
```

```python
# Creating X Vector
X = df3.drop("price", axis="columns")
X.head()
```

```python
X.shape
```

```python
# Creating y
y = df3["price"]
y.head()
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
# Now let Create a gradient descent for our data set
# Gradient Descent Function
def gradient_descent(X, y, w, b, alpha, iterations):
    x_means = X.mean()
    X = X / x_means

    n = len(y)
    costs = []  # To store costs for each iteration
    
    
    i = 0
    while i<=iterations:
        y_predict = np.dot(X, w) + b

        # Calculate the cost (Mean Squared Error) for the current predictions
        cost = (1 / (2 * n)) * np.sum((y_predict - y) ** 2)

        
        # Calculate gradients for weights and bias
        dw = (1/n) * np.dot(X.T, (y_predict - y))
        db = (1/n) * np.sum(y_predict - y)
        
        # Update weights and bias using the gradients and learning rate
        w = w - alpha * dw
        b = b - alpha * db
        
#         print(f"cost={cost}, i={i}")
        
        # Append the current cost to the costs list for visualization
        costs.append(cost)
        i+=1

    
    w = w/x_means
    return w, b, costs, i
```

```python
rows, cols = X_train.shape
w = np.zeros(cols)
b = 0
alpha = 0.0006
iterator = 10000 # If i will increase the number of iteration then it will give us more accurate ans

w_updated, b_updated, costs, iterations = gradient_descent(X_train, y_train, w, b, alpha, iterator)
```

```python
w_updated, b_updated
```

```python
def predict(X, w, b):
    return np.dot(X, w)+b
```

```python
# Now finding our R Squared to see the goodness of our eqn
y_pred = np.dot(X_train, w_updated) + b_updated
SSR = np.sum((y_pred - y_train)**2)
SST = np.sum((y_train - y_train.mean())**2)
Rsqr = 1 - (SSR / SST)
Rsqr # Range: [1-0]
```

```python
# Graph of learning rate
plt.plot(range(iterations), costs, label="learning rate")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Function Evolution")
plt.legend()
plt.show()
```
